```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load dataset

def load_dataset(file_path):
  return pd.read_csv(file_path)

df = load_dataset("/content/drive/MyDrive/Colab Notebooks/Health_heart_experimental.csv")
df.head(5)
```

line chart

```python
line = df.groupby('age').mean(numeric_only=True).reset_index()
plt.figure(figsize=(14, 6))
# line chart
sns.lineplot(x='age',y='SysBP',data=line)
```

bar plot



```python
plt.figure(figsize=(14, 6))
sns.barplot(x='age',y='HR',hue='sex',data=df)
```

pie chart




```python
df['sex'].value_counts().plot(kind='pie',autopct='%0.1f')
```

box plot



```python
sns.boxplot(df['SysBP'])
```

histogram


```python
plt.figure(figsize=(20, 6))
plt.hist(df['DiaBP'])
```

scatter plot


```python
plt.figure(figsize=(14, 6))
sns.scatterplot(x='age',y='BMI',hue='sex',data=df)
plt.show()
```
